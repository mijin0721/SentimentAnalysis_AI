```python
from transformers import AutoTokenizer, AutoModel

tokenizer = AutoTokenizer.from_pretrained("microsoft/xtremedistil-l12-h384-uncased")

model = AutoModel.from_pretrained("microsoft/xtremedistil-l12-h384-uncased")
```

```python
from torch import nn

class MyBert(nn.Module):
    def __init__(self, tokenizer, model):
        super(MyBert, self).__init__()
        self.tokenizer = tokenizer
        self.bert = model
        self.classifier = nn.Linear(384, 2)
    
    def forward(self, x):
        out = self.bert(**self.tokenizer(x, return_tensors="pt"))
        out = self.classifier(out)
        return out
```

```python
mymodel = MyBert(tokenizer, model)
```

```python
from transformers import TextClassificationPipeline, BertForSequenceClassification, AutoTokenizer

model_name = 'smilegate-ai/kor_unsmile'
model = BertForSequenceClassification.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
model.classifier = nn.Linear(768, 1)
```

```python
pipe = TextClassificationPipeline(
        model = model,
        tokenizer = tokenizer,
        device = -1,   # cpu: -1, gpu: gpu number
        return_all_scores = True,
        function_to_apply = 'sigmoid'
    )
```
